```python
import pandas as pd
mutations = pd.read_csv('summary.hg19_multianno.csv')
mutations.shape
```

```python
exonic = mutations[mutations['Func.refGene'] == 'exonic']
exonic.head()
exonic.shape
```

```python
all_gene = exonic['Gene.refGene'].value_counts()
all_gene = all_gene.index.tolist()
```

```python
nonsyn = exonic[exonic['ExonicFunc.refGene'] == 'nonsynonymous SNV']
nonsyn.shape
nonsyn['Gene.refGene'].value_counts().shape
```

```python
common = exonic[exonic['Sample'].str.contains(':')]
common
common_list = common['Sample'].tolist()
common_list_1 = [i.split(':')[0] for i in common_list]
common_list_2 = [i.split(':')[1] for i in common_list]
common['Sample_list_1'] = common_list_1
common['Sample_list_2'] = common_list_2
common.head()
common_1 = common.drop('Sample_list_2',axis = 1, inplace=False)
common_1['sample_list'] = common_1['Sample_list_1']
common_1.drop('Sample_list_1',axis = 1,inplace = True)
common_2 = common.drop('Sample_list_1', axis = 1,inplace = False)
common_2['sample_list'] = common_2['Sample_list_2']
common_2.drop('Sample_list_2',axis = 1, inplace = True)
common_2
common_split = common_1.append(common_2)
common_split['Sample'] = common_split['sample_list']
common_split.drop('sample_list',axis = 1,inplace = True)
common_split
```

```python
exonic_split = exonic.drop(common.index, axis = 0, inplace=False)
exonic_split = exonic_split.append(common_split)
exonic_split.shape
```

```python
def clean(df):
    counts = pd.DataFrame()
    for item in all_gene:
        data = df[df['Gene.refGene'] == item]
        count = pd.DataFrame(data['Sample'].value_counts())
        count.columns = [item]
        counts = pd.concat([counts,count],axis = 1)
        counts = counts.fillna(0)
    return counts
cleaned_exonic = clean(exonic_split)
cleaned_exonic_t = cleaned_exonic.T 
```

```python
cleaned_exonic_t.index
cleaned_exonic_t['Gene'] = cleaned_exonic_t.index
col = cleaned_exonic_t.columns.tolist()
col = col[-1:] + col[:-1]
cleaned_exonic_t=cleaned_exonic_t[col]
cleaned_exonic_t = cleaned_exonic_t.fillna(0)
```

```python
exonic_split['ExonicFunc.refGene'].value_counts()
sample_list = exonic_split['Sample'].value_counts().index.tolist()
def count(df):
    counts = pd.DataFrame()
    for item in sample_list:
        data = df[df['Sample'] == item]
        count = pd.DataFrame(data['ExonicFunc.refGene'].value_counts())
        count.columns = [item]
        counts = pd.concat([counts,count],axis = 1)
        counts = counts.fillna(0)
    return counts
syn_count = count(exonic_split)
syn_count = syn_count.T
syn_count['nonsyn'] = syn_count['nonsynonymous SNV']+syn_count['stopgain']+syn_count['stoploss']
syn_count['total'] = syn_count['nonsyn']+syn_count['synonymous SNV']
```

```python
syn_count['Sample'] = syn_count.index
syn_count.sort_values('Sample', inplace = True)
syn_count.head(3)
```

```python
f, ax = plt.subplots(figsize=(20, 5))
sns.set_color_codes("muted")

sns.barplot(x="Sample", y="total", data=syn_count,
            label="Synonymous", color="g")
sns.set_color_codes("muted")

sns.barplot(x="Sample", y="nonsyn", data=syn_count,
            label="Nonsynonymous", color="b")

ax.legend(ncol=5, loc="upper right", frameon=True,prop={'size':16})
ax.set(xlabel="Sample ID",
       ylabel="")
ax.set_xlabel(xlabel = 'Sample ID',fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=20)
plt.xticks(rotation=70)
sns.despine(left=True, bottom=True)
```

```python
exonic_split.head()
synon = exonic_split[(exonic_split['ExonicFunc.refGene'] == 'nonsynonymous SNV')|(exonic_split['ExonicFunc.refGene'] == 'stopgain')|(exonic_split['ExonicFunc.refGene'] == 'stoploss')]
synon_data = synon[['Sample','Gene.refGene','ExonicFunc.refGene']]
synon_data
```

```python
synon_data['ExonicFunc.refGene'].value_counts()
```

```python
synon_data['ExonicFunc.refGene'].replace('nonsynonymous SNV','1',inplace = True)
synon_data['ExonicFunc.refGene'].replace('stopgain','100',inplace = True)
synon_data['ExonicFunc.refGene'].replace('stoploss','100',inplace = True)
synon_data['ExonicFunc.refGene'] = synon_data['ExonicFunc.refGene'].astype('int')
synon_data
```

```python
import numpy as np
synon_pivot = synon_data.pivot_table(index = 'Gene.refGene',columns = 'Sample',values = 'ExonicFunc.refGene',aggfunc=np.sum)
columns = sample_list
synon_pivot[columns] = synon_pivot[columns].astype('float')
synon_pivot = synon_pivot.fillna('0')
synon_pivot_t = synon_pivot.T
synon_pivot_t_highfreq = synon_pivot_t[cols]
synon_pivot_t_highfreq = synon_pivot_t_highfreq.T
synon_pivot_t_highfreq
s_c = synon_pivot_t_highfreq.columns
synon_pivot_t_highfreq[s_c] = synon_pivot_t_highfreq[s_c].astype('float')
```

```python
def change(x):
    if x >=100:
        return 2
    elif 0<x<100:
        return 1
    else:
        return x
synon_pivot_t_highfreq = synon_pivot_t_highfreq.applymap(change)
synon_pivot_t_highfreq
```

```python
f, ax = plt.subplots(figsize=(20, 5))
sns.heatmap(synon_pivot_t_highfreq,cbar=False,square = True,cmap="Blues",linewidths=.5)
```
